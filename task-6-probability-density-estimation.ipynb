```r
data <- read.csv('/kaggle/input/sa2-statistical-dataset-samples/Salary_Data.csv')

head(data)
```

```r
salary <- data$Salary
```

```r

# Histogram-based density estimation
hist(salary, 
     breaks = "Sturges",   
     probability = TRUE, 
     col = "lightblue", 
     border = "white", 
     main = "Histogram-Based Density of Salary",
     xlab = "Salary")

# Add empirical density estimate
lines(density(salary), col = "blue", lwd = 2)

# theoretical density
x_vals <- seq(min(salary), max(salary), length.out = 100)
y_vals <- dnorm(x_vals, mean = mean(salary), sd = sd(salary))
lines(x_vals, y_vals, col = "red", lwd = 2, lty = 2)

# legends (blue for empirical and red for theoretical)
legend("topright", legend = c("Empirical Distribution", "Theoretical Distribution"),
       col = c("blue", "red"), lwd = 2, lty = c(1, 2))
```

#### Visualization

##### Histogram

The x-axis in the plot represents the salary, and the y-axis in the plot represents the density, not the raw counts, meaning this represents the proportion of data within each salary bin.

##### Empirical Distribution

The blue line in the plot represents the empirical distribution, which is the estimates of the salary in the actual salary distribution in the salary dataset. In this smoothed line (blue line), we can observe that there are two humps in the plot; the first one can be seen at around 60,000, and the second hump can be seen at 110,000. Since the empirical line has two humps, this indicates that we have a bimodal distribution of salary.

##### Theoretical Distribution 

The red dashed curve line represents the theoretical distribution, which assumes a normal distribution based on the salary dataset mean and standard deviation. The appearance of the curve line suggests what a salary distribution would look like if salaries followed the normal distribution.

##### Comparison (Empirical vs. Theoretical)

When it comes to the shape of the line, the empirical reflects the real distribution (bimodal, skewed), while the theoretical assumes normal distribution (unimodal and symmetric). When it comes to data dependency, the empirical follows the actual data, while the theoretical follows the standard deviation and mean of the dataset.

In conclusion, the empirical distribution (KDE) gives us an accurate and flexible representation of the salary data. Histogram-based estimation + empirical distribution is the best method here we have in our dataset when the normality cannot be assumed.
